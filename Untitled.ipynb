```python
import pandas as pd
import re
```

```python
cat = pd.read_csv('obel-p.tsv', encoding='utf8', sep='\t').fillna('')
cat
```

```python
cat['museum_no2'] = [re.sub(' 0+', ' ', no) for no in cat.museum_no]
cat['designation2'] = [re.sub(' 0+', ' ', des) for des in cat.designation]
```

```python
cat['museum_no3'] = [' '.join(no.split()[:2]) for no in cat.museum_no2]
```

```python
cat[['designation', 'id_text', 'museum_no', 'museum_no2', 'museum_no3', 'designation2']][:10]
```

```python
dict_desig = dict(zip(cat['designation2'], cat['id_text']))
dict_musno = dict(zip(cat['museum_no2'], cat['id_text']))
dict_musno_short = dict(zip(cat['museum_no3'], cat['id_text']))
```

```python
dict_musno_short
```

```python
with open('Delnero_main.txt', 'r', encoding='utf8') as d:
    lit = d.read().splitlines()
```

```python
Pnos = []
n = 0
for line in lit:
    if line[:2] == '&&':
        l = line.split()
        l = ' '.join(l[:2])
        l = l[2:]
        p = dict_musno_short.get(l, l)
        if p == l:
            p = dict_desig.get(l, l)
            if p == l:
                p = 'unknown'
                n = n+1
        q = [p, l]
        Pnos.append(q)
print(str(len(Pnos)) + ' ; ' + str(n) + ' unknown')
```
